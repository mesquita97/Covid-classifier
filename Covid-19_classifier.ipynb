* ## Pre-processing images

On the next part of the code I'll apply some preprocessing image techniques to achieve a good classifier (avoid overfitting for example), the ImageDataGenerator function applies all the processing that you define in its arguments before  fitting the model, also with the flow_from_directory function it'll splits the data in train and test samples according to their respective directories.

```python
# Defining some constant variables to use in the preprocessor

train_path  = '../working/train'
test_path =  '../working/test'
IMAGE_SIZE    = (250, 250) # 150x150 (height, width) pixels
NUM_CLASSES   = 3 #COVID-19, NORMAL and PNEUMONIA
BATCH_SIZE    = 32  # try reducing batch size or freeze more layers if your GPU runs out of memory
NUM_EPOCHS    = 20 # idem for epochs
```

```python
#Train datagen here is a preprocessor
train_datagen = ImageDataGenerator(rescale=1./255,
                                   horizontal_flip = False ,
                                   vertical_flip = False ,
                                   validation_split = 0.2,
                                  data_format='channels_last')

train_batches = train_datagen.flow_from_directory(train_path,
                                                  target_size=IMAGE_SIZE,
                                                  shuffle=True,
                                                  subset = "training",
                                                  class_mode="categorical"
                                                  )

valid_batches = train_datagen.flow_from_directory(train_path,
                                                  target_size=IMAGE_SIZE,
                                                  shuffle=True,
                                                  subset = "validation",
                                                  class_mode="categorical"
                                                  )
```

* ## Fitting the models

To build a new architeture model is appropiate to base on the most powerful already existing models.
Let's fit three models based on the LeNet-5 and VGG architecturies, and then we'll obtain which of them works better for this dataset.

The two firsts models were based on the LeNet-5. They're a Conv-Pool-Conv-Pool kind of architecture.

* ### First Model

```python
### Model 1
def fit_model_1():
    
    # define model
    model = Sequential()
    
    #Convolve block
    model.add(Conv2D(8, kernel_size=(3, 3), activation='relu',padding='same',input_shape=(250,250, 3)))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Conv2D(16, kernel_size=(5, 5), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Flatten())
    model.add(Dense(120, activation='relu'))
    model.add(Dense(60, activation='relu'))
    
    #Output layer
    model.add(Dense(3, activation='softmax'))
    
    # compile model
    opt = Adam(lr=0.001)
    model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    
    #Defining callbacks
    early_stop = EarlyStopping(patience=3, monitor='val_accuracy')
    
    # fit model
    STEP_SIZE_TRAIN=train_batches.n//train_batches.batch_size
    STEP_SIZE_VALID=valid_batches.n//valid_batches.batch_size
    result=model.fit_generator(train_batches,
                        validation_data = valid_batches,
                        callbacks=[early_stop],
                        epochs = NUM_EPOCHS                        
                       )
    return model
```

* ### Second Model

The difference for the second model is that I double the value of the filters for each next layer

```python
def get_model_2():
    # define model
    model = Sequential()
    
    #Convolve block
    model.add(Conv2D(32, kernel_size=(3, 3), activation='relu',input_shape=(250, 250, 3)))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    
    #Output layer
    model.add(Dense(3, activation='softmax'))
    
    # compile model
    opt = Adam(lr=0.001)
    model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    
    #Defining callbacks
    early_stop = EarlyStopping(patience=3, monitor='val_accuracy')
    
    # fit model
    STEP_SIZE_TRAIN=train_batches.n//train_batches.batch_size
    STEP_SIZE_VALID=valid_batches.n//valid_batches.batch_size
    result=model.fit_generator(train_batches,
                        validation_data = valid_batches,
                        validation_steps = STEP_SIZE_VALID,
                        callbacks=[early_stop],
                        epochs = NUM_EPOCHS                        
                       )
    return model
```

```python
def get_model_3():
    # define model
    model = Sequential()
    
    model.add(Conv2D(16, kernel_size=(3, 3), activation='relu',padding='same',input_shape=(250, 250, 3)))
    model.add(Conv2D(16, kernel_size=(3, 3), activation='relu',padding='same'))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
    model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    
    #Output layer
    model.add(Dense(3, activation='softmax'))
    
    # compile model
    opt = Adam(lr=0.001)
    model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    
    #Defining callbacks
    early_stop = EarlyStopping(patience=2, monitor='val_accuracy')
    
    # fit model
    STEP_SIZE_TRAIN=train_batches.n//train_batches.batch_size
    STEP_SIZE_VALID=valid_batches.n//valid_batches.batch_size
    result=model.fit_generator(train_batches,
                        steps_per_epoch =STEP_SIZE_TRAIN,
                        validation_data = valid_batches,
                        validation_steps = STEP_SIZE_VALID,
                        callbacks=[early_stop],
                        epochs = NUM_EPOCHS                        
                       )
    return model
```

* ## Fitting the new Model

In this section I'll fit the new model created by me inspired on the two better models that I've tested above.

To get a better perfomance of this new model I've maintened a similar structure of Conv-Pool-Conv-Pool, but choosing different attributes while I was getting better accuracies (val_accuracy). 

It was created another train_datagen that contains more preprocessing techniques aiming to avoid overfitting.

```python
#Train datagen here is a preprocessor
train_datagen = ImageDataGenerator(rescale=1./255,
                                   horizontal_flip = False ,
                                   vertical_flip = False ,
                                   validation_split = 0.2,
                                   zca_whitening=True,
                                   zoom_range=0.1,
                                   data_format='channels_last')

train_batches = train_datagen.flow_from_directory(train_path,
                                                  target_size=IMAGE_SIZE,
                                                  shuffle=True,
                                                  subset = "training",
                                                  class_mode="categorical"
                                                  )

valid_batches = train_datagen.flow_from_directory(train_path,
                                                  target_size=IMAGE_SIZE,
                                                  shuffle=True,
                                                  subset = "validation",
                                                  class_mode="categorical"
                                                  )
```

```python
# Defining the model

def get_refine_model():
    # define model
    model = Sequential()
    
    #Convolve block
    model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same', input_shape=(250, 250, 3)))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(BatchNormalization())
    model.add(Dropout(0.6))
    
    model.add(Conv2D(128, kernel_size=(5, 5), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Dropout(0.3))
    model.add(Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Dropout(0.3))
    model.add(Conv2D(256, kernel_size=(5, 5), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
    model.add(Dropout(0.3))
   
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dense(32, activation='relu'))
    
    #Output layer
    model.add(Dense(3, activation='softmax'))
    
    # compile model
    opt = Adam(lr=0.001)
    model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    
    #Defining callbacks
    checkpoint = ModelCheckpoint(filepath='COVID_classifier.hdf5', save_best_only=True, save_weights_only=True)
    
    # fit model
    STEP_SIZE_TRAIN=train_batches.n//train_batches.batch_size
    STEP_SIZE_VALID=valid_batches.n//valid_batches.batch_size
    result=model.fit_generator(train_batches,
                        steps_per_epoch =STEP_SIZE_TRAIN,
                        validation_data = valid_batches,
                        validation_steps = STEP_SIZE_VALID,
                        callbacks=[checkpoint],
                        epochs = 25                                          
                       )
    return model, result
```

```python
# Fitting the model

classifier, result = get_refine_model()
```

```python
def plot_accs(result, epochs):
    acc = result.history['accuracy']
    loss = result.history['loss']
    val_acc = result.history['val_accuracy']
    val_loss = result.history['val_loss']
    plt.figure(figsize=(15, 5))
    plt.subplot(121)
    plt.plot(range(1,epochs), acc[1:], label='Train_acc')
    plt.plot(range(1,epochs), val_acc[1:], label='Test_acc')
    plt.title('Accuracy over ' + str(epochs) + ' Epochs', size=15)
    plt.legend()
    plt.grid(True)
    plt.subplot(122)
    plt.plot(range(1,epochs), loss[1:], label='Train_loss')
    plt.plot(range(1,epochs), val_loss[1:], label='Test_loss')
    plt.title('Loss over ' + str(epochs) + ' Epochs', size=15)
    plt.legend()
    plt.grid(True)
    plt.show()
    
plot_accs(result, 25)
```

## Predicting

I'll use the saved model that was evaluated above to predict the pacient conditions with the unseen data in the test folder.

```python
# Evaluating the model

eval_generator.reset()  
x = classifier.evaluate_generator(eval_generator,
                           steps = np.ceil(len(eval_generator) / 1),
                           use_multiprocessing = False,
                           verbose = 1,
                           workers=1
                           )


print('Test loss:' , x[0])
print('Test accuracy:',x[1])
```

```python
scores = classifier.predict_generator(eval_generator)
y_pred=np.argmax(scores, axis=1)
```
